Iterate through a list

```python
fruits = ["apple","banana","mango"]
for fruit in fruits:
  print(fruits)
```

```python
count = 1  #while loop
while count <= 9:
  print(count)
  count += 1
```

```python
for i in range(1,8):  #break - stops the loops immediately.
  if i == 5:
    break
  print(i)
```

```python
for i in range(1,8):  #continue - skips the current iteration.
  if i == 5:
    continue
  print(i)
```

```python
for i in range(1,8):  #pass - placeholder statement
  if i == 5:
    pass
  print(i)
```

```python
number = int(input("enter a number ")) #question 1
sum = 0
for i in range(1, number + 1):
    sum = sum + i

print("Sum =", sum)
```

```python
correct_password = "ankush"  #question 2
entered_password = ""

while entered_password != correct_password:
  entered_password = input("enter correct password: ")
  if entered_password != correct_password:
    print("Incorrect password. Please try again.")

print("Correct password!")
```

```python
n = int(input("Enter number of elements: ")) #question 3

count = 0

for i in range(n):
    num = int(input("Enter number: "))
    if num > 0:
        count += 1

print("Positive numbers =", count)
```

```python
sales = float(input("Enter monthly sales: ")) #question 4
target = float(input("Enter target sales: "))

if sales >= target:
    print("Target Achieved")
else:
    print("Target Not Achieved")
```
